### Global available data

```python
# Getting the data from the top 10 expensive neighbourhoods
to_neigh_df=to_data.reset_index(drop=True)
neigh_avg_data = to_neigh_df.groupby('neighbourhood').mean()
top_10_neigh_data = neigh_avg_data.nlargest(10, "average_house_value")
top_10_neigh_data.reset_index(inplace=True)

# Calculate the mean number of dwelling types units per year
to_dwelling_types = to_data.drop(columns=['average_house_value', 'shelter_costs_owned',
                                          'shelter_costs_rented']).groupby(to_data.index).sum() 
# Calculate the average monthly shelter costs for owned and rented dwellings
to_avg_shelter_cost = to_data[["shelter_costs_owned","shelter_costs_rented"
                    ]].groupby(to_data.index).mean() 
# Calculate the average house value per year
to_avg_house_val=to_data[["average_house_value"]].groupby(to_data.index).mean() 
```

### Panel Visualization Functions

```python
# Define Panel visualization functions
def neighbourhood_map():
    df_neighbourhood_locations = pd.read_csv(file_path)
    to_neigh_df=to_data.reset_index(drop=True)
    mean_neigh_val_df=to_neigh_df.groupby('neighbourhood').mean()
    mean_neigh_val_df.reset_index(inplace=True)
    df_neighbourhood_index = df_neighbourhood_locations.set_index("neighbourhood")
    mean_neigh_val_index = mean_neigh_val_df.set_index("neighbourhood")
    avg_val_by_loc = pd.concat([df_neighbourhood_index , mean_neigh_val_index], axis=1).reset_index()
    to_plot = px.scatter_mapbox(
        avg_val_by_loc,
        lat="lat",
        lon="lon",
        hover_data = avg_val_by_loc,
        color="average_house_value",
        size="average_house_value",
        color_continuous_scale=px.colors.cyclical.IceFire,
        title="Average House Values in Toronto",
        width=1200, height = 500
   )
    return to_plot
       
def create_bar_chart(data, title, xlabel, ylabel, color):
    fig = data.plot(kind = "bar", color = color)
    fig.set_title(title)
    fig.set_xlabel(xlabel)
    fig.set_ylabel(ylabel)   
    return fig.figure

def create_line_chart(data, title, xlabel, ylabel, color):
    chart = data.plot(color = color)
    chart.set_title(title)
    chart.set_xlabel(xlabel)
    chart.set_ylabel(ylabel)
    return chart.figure
    
def average_house_value():
    ax = to_avg_house_val.plot(legend = False) 
    ax.set_title("Average House Value in Toronto")
    ax.set_xlabel("Year")
    ax.set_ylabel("Avg. House Value")
    return ax.figure
    
def average_value_by_neighbourhood():
    avg_neigh_house_val = to_data[["neighbourhood","average_house_value"]]            
    avg_neigh_house_val.reset_index(inplace=True)
    return avg_neigh_house_val.hvplot.line(x="year", y="average_house_value", groupby="neighbourhood", 
                           xlabel="Year", ylabel="Avg. House Value").opts(yformatter="%.0f")
    
def number_dwelling_types():
    to_dwell_types=to_data[["neighbourhood","single_detached_house","apartment_five_storeys_plus",
                        "movable_dwelling", "semi_detached_house", "row_house", "duplex",
                        "apartment_five_storeys_less","other_house"]]
    to_dwell_types.reset_index(inplace=True)
    return to_dwell_types.hvplot.bar(x="year", y=("single_detached_house","apartment_five_storeys_plus","movable_dwelling",  "semi_detached_house",
                          "row_house", "duplex",  "apartment_five_storeys_less","other_house"),
                          groupby="neighbourhood", xlabel="Year", ylabel="Dwelling Type Units", 
                          width=800, height = 500, rot=90) 
        
def average_house_value_snapshot():
    to_house_val_yearly = to_data[["neighbourhood","average_house_value"]]
    to_house_val_yearly.reset_index(inplace=True)
    return px.bar(to_house_val_yearly, x="neighbourhood", y="average_house_value", 
                  color= "average_house_value",facet_row="year")
    
def top_most_expensive_neighbourhoods():
    plot=top_10_neigh_data.hvplot.bar(x="neighbourhood", y="average_house_value", title="Top 10 Expensive Neighbourhoods in Toronto",
                            xlabel="Neighbourhood", ylabel= "Avg. House Value", width=800, height = 500, rot=90).opts(yformatter="%.0f")
    return plot
    
def sunburts_cost_analysis():
    top_neigh_yr=to_data.reset_index()
    top_10_neigh_yr=top_neigh_yr.sort_values(["year","average_house_value"], 
                                             ascending=[True, False]).groupby("year").head(10)
    return px.sunburst(top_10_neigh_yr, path=['year', 'neighbourhood'], values="shelter_costs_owned",
            color='shelter_costs_owned', color_continuous_scale='blues', 
            hover_data=["average_house_value",'shelter_costs_rented'], 
            title= "Cost Analysis of Most Expensive Neighbourhoods in Toronto per year")
    
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create a Title for the Dashboard
title = pn.pane.Markdown("# Real Estate Analysis of Toronto from 2001 to 2016")    

# Define a welcome text
text = pn.pane.Markdown("This dashboard presents a visual analysis of historical house values, dwelling types per neighbourhood and dwelling costs in Toronto, Ontario according to census data from 2001 to 2016.\
You can navigate through the tabs above to explore more details about the evolution of the real estate market on The 6 across these years.")          

# Create a tab layout for the dashboard
welcome = pn.Column(title, text, neighbourhood_map)
yearly_mrkt_row = pn.Row(plot_2001, plot_2006)
yearly_mrkt_row_1 = pn.Row(plot_2011, plot_2016)
yearly_mrkt_analysis=pn.Column(yearly_mrkt_row, yearly_mrkt_row_1 )
shelter_house_val = pn.Column(shelter_cost_own, shelter_cost_rent, average_house_value)
neighbourhood_row = pn.Row(average_value_by_neighbourhood, average_house_value_snapshot )
neighbouhood_analysis = pn.Column(neighbourhood_row, number_dwelling_types) 
top_neighbourhoods =  pn.Row(top_most_expensive_neighbourhoods, sunburts_cost_analysis)

# Create the main dashboard
to_real_estate_mrkt = pn.Tabs(
    ("Welcome", welcome),
    ("Yearly Market Analysis", yearly_mrkt_analysis),
    ("Shelter Costs Vs. House Value", shelter_house_val),
    ("Neighbourhood Analysis", neighbouhood_analysis),
    ("Top Expensive Neighbourhoods", top_neighbourhoods))
to_real_estate_mrkt
```
